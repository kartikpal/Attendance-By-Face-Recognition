```python
from tkinter import * 
import tkinter as tk
from tkinter import ttk
from PIL import Image,ImageTk
import import_ipynb

import Student
import os

class Face_Recognition_System:
    def __init__(self,root):
        self.root=root
        self.root.geometry("1500x790+0+0")
        self.root.title("Face Recognition System")
        self.root.resizable(False,False)
        
        img=Image.open("2.png")
       
        self.photoimg=ImageTk.PhotoImage(img)
        #makes the variable an instance variable and it will not be garbage collected.
        
        f_lbl=Label(self.root,image=self.photoimg)
        f_lbl.place(x=0,y=0,width=1500,height=130)
        
        #bg image 
        img2=Image.open("3.png")
        img2=img2.resize((1550,710),Image.ANTIALIAS)
        self.photoimg2=ImageTk.PhotoImage(img2)
        #makes the variable an instance variable and it will not be garbage collected.
        
        bg_img=Label(self.root,image=self.photoimg2)
        bg_img.place(x=0,y=130,width=1550,height=750)
        
        #title
        title_lbl=Label(bg_img,text= "FACE RECOGNITION ATTENDANCE SYSTEM",font=("times new roman",35,"bold"),bg="white",fg="red")
        title_lbl.place(x=0,y=0,width=1600,height=45)
        
        #student button
        img4=Image.open("4.png")
        self.photoimg4=ImageTk.PhotoImage(image=img4)
        
        b1=Button(bg_img,image=self.photoimg4,cursor="hand2",command=self.student_details)
        b1.place(x=200,y=100,width=220,height=220)
        
        b2=Button(bg_img,text="Student Details",cursor="hand2",font=("times new roman",15,"bold"),command=self.student_details)
        b2.place(x=200,y=300,width=220,height=40)
        
        #detect face button
        
        img5=Image.open("5.png")
        self.photoimg5=ImageTk.PhotoImage(image=img5)
        
        b3=Button(bg_img,image=self.photoimg5,cursor="hand2")
        b3.place(x=500,y=100,width=220,height=220)
        
        b4=Button(bg_img,text="Face Detection",cursor="hand2",font=("times new roman",15,"bold"))
        b4.place(x=500,y=300,width=220,height=40)
        
        
       #Attendance
    
        img6=Image.open("8.png")
        self.photoimg6=ImageTk.PhotoImage(img6)
        
        b5=Button(bg_img,image=self.photoimg6,cursor="hand2")
        b5.place(x=800,y=100,width=220,height=220)
        
        b6=Button(bg_img,text="Attendance",cursor="hand2",font=("times new roman",15,"bold"))
        b6.place(x=800,y=300,width=220,height=40)
        
       
    #train data
    
        img7=Image.open("7.png")
        self.photoimg7=ImageTk.PhotoImage(img7)
        
        b7=Button(bg_img,image=self.photoimg7,cursor="hand2")
        b7.place(x=1100,y=100,width=220,height=220)
        
        b8=Button(bg_img,text="Train Data",cursor="hand2",font=("times new roman",15,"bold"))
        b8.place(x=1100,y=300,width=220,height=40)
        
    #Photos
    
        img8=Image.open("9.png")
        self.photoimg8=ImageTk.PhotoImage(img8)
        
        b9=Button(bg_img,image=self.photoimg8,cursor="hand2",command=self.open_img)
        b9.place(x=500,y=380,width=220,height=220)
        
        b10=Button(bg_img,text="Photos",cursor="hand2",font=("times new roman",15,"bold"),command=self.open_img)
        b10.place(x=500,y=580,width=220,height=40)
    #Exit
        img9=Image.open("10.png")
        self.photoimg9=ImageTk.PhotoImage(img9)
        
        b9=Button(bg_img,image=self.photoimg9,cursor="hand2")
        b9.place(x=800,y=380,width=220,height=220)
        
        b10=Button(bg_img,text="Exit",cursor="hand2",font=("times new roman",15,"bold"))
        b10.place(x=800,y=580,width=220,height=40)
        
        def open_img(self):
            os.startfile("data")
            
            
        #========function====#
        
        def student_details(self):
            self.new_window=Toplevel(self.root)
            self.app=Student(self.new_window)
        
if __name__=="__main__":
    root=Tk()
    obj=Face_Recognition_System(root)
    root.mainloop()        
```
